```python
# as we observe in gender column there are (m,male,Male) and (f,female,Female)

df['Gender'] = df['Gender'].replace({
    'M':'Male',
    'F':'Female',
    'male':'Male',
    'female':'Female'
})
```

```python
df
```

```python
# in cabin column there are more speces

df['Cabin_Class'] = df['Cabin_Class'].str.strip()
```

```python

df[df['Age']< 0]
```

```python
df[df['Age']> 100]
```

```python
# there are age columns which are greater than 100 and less than 0 

df.loc[df['Age']< 0,'Age'] = np.nan
```

```python
df.loc[df['Age']> 100,'Age'] = np.nan
```

```python
# from the above we fixed the invalid ages now we need to fill them
df['Age'].fillna(df['Age'].median)
```

```python
df['Age_Group'] = pd.cut(df['Age'],bins=[0,18,35,50,100],
                         labels=['Teen','Young Adult','Adult','Senior'])
```

```python
# Ticket fare has some in rupess,dolars,USD

df['Ticket_Fare'] = (
    df['Ticket_Fare']
    .astype(str)
    .str.replace('₹','')
    .str.replace(',','')
    .str.replace('USD','')
    .astype(float)
)
```

```python
df['Ticket_Fare']
```

```python
df['Revenue_Category'] = pd.cut(df['Ticket_Fare'],bins=[0,5000,15000,50000],
labels=['Low','Medium','High'])
```

```python
plt.figure(figsize=(6,3))
sns.boxplot(x=df['Ticket_Fare'])
plt.show()
```

```python
Q1 = df['Ticket_Fare'].quantile(0.25)
Q3 = df['Ticket_Fare'].quantile(0.75)

IQR = Q3 - Q1
lower = Q1 - 1.5*IQR
upper = Q3 + 1.5 * IQR

df = df[
(df['Ticket_Fare'] >= lower) &
(df['Ticket_Fare'] <= upper)
]
```

```python
sns.countplot(
    x = 'Passenger_Satisfaction',
    data = df
)
plt.show()
```

```python
# Cabin Class Analysis

df['Cabin_Class'].value_counts()
```

```python
# Cabin Class vs Satisfaction

sns.countplot(
    x='Cabin_Class',
    hue= 'Passenger_Satisfaction',
    data = df
)
```

```python
# Revenue by Cabin Class

df.groupby("Cabin_Class")['Ticket_Fare'].sum()
```

### Final Business Conclusions
##### Key Findings
